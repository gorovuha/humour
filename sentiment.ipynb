```python
import pandas as pd
from transformers import pipeline
```

```python
data = pd.read_csv('dub0_5.csv')
```

```python
data['full_joke'] = data['set-up']+data['punchline']
data['label'] = ['' for i in range(len(data))]

```

```python
labels = []

classifier = pipeline("zero-shot-classification", model="MoritzLaurer/DeBERTa-v3-large-mnli-fever-anli-ling-wanli")
candidate_labels = ["politics", "neutral", "racist", "offending"]
for i in range(len(data['full_joke'])):
    output = classifier(data['full_joke'][i], candidate_labels, multi_label=False)
    labels.append(output['labels'][output['scores'].index(max(output['scores']))])
```

```python
data['label'] = labels
```

```python
data.to_csv('labeled.csv')
```
